```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('clean_data')
df.head()
```

```python
df.shape
```

```python
grouped_df = df.groupby('production_year')
grouped_df.head()
```

```python
production_by_year = df.groupby(['production_year', 'movie'])['production_budget'].mean()
production_by_year
```

```python
sns.lineplot(x='production_year', y='production_budget', legend='brief', data=df).set_title('Production Budget by Year')
#Question 1: Has the average cost for production increased, decreased or remained the same throught the years?
#The average production cost of movies has been increaseing throughout the years.
#If microsoft is considering long term production projects, it is important to keep in mind the production cost of 
#future projects will be more.
```

```python
year_df = df.set_index('production_year')
```

```python
year_df = year_df.sort_index(ascending = True)
year_df
```

```python
grouped_df = df.groupby(['production_year'])['production_budget', 'total_revenue'].mean()
grouped_df
```

```python
sns.lineplot(data=grouped_df).set_title('Production Budget and Total Revenue by Year')
#Question 2: What is the corrolation between average production budget and total revenue?
#The average total revenue throughout the years has also increased with production cost. 
#This direct corolation could be possibly just be because of inflation in general costs.
```
