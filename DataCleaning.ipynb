```python
#clean second dataframe
budgets_df = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
print(len(tn_movie_budgets_df))
budgets_df.head()
```

```python
budgets_df['production_year'] = budgets_df.release_date.map(lambda x: x[7:])
budgets_df.head() #add column of year movie realeased
```

```python
budgets_df['production_budget'].dtype
```

```python
budgets_df['production_budget'] = budgets_df['production_budget'].str.replace('$', '')
budgets_df['production_budget'] = budgets_df['production_budget'].str.replace(',', '')
#change budget strings to integers
```

```python
budgets_df['domestic_gross'] = budgets_df['domestic_gross'].str.replace('$', '')
budgets_df['domestic_gross'] = budgets_df['domestic_gross'].str.replace(',', '')
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].str.replace('$', '')
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].str.replace(',', '')

budgets_df['domestic_gross'] = budgets_df['domestic_gross'].astype('int64')
budgets_df['worldwide_gross'] = budgets_df['worldwide_gross'].astype('int64')
```

```python
budgets_df
```

```python
budgets_df['total_revenue'] = budgets_df['domestic_gross'] + budgets_df['worldwide_gross']
budgets_df.head()
```

```python
budgets_df.groupby(['production_year'])['production_budget', 'total_revenue'].mean()
```

```python
budgets_df.drop(budgets_df[budgets_df['total_revenue'] == 0].index, inplace=True)
```
